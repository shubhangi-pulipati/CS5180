```python
import matplotlib.pyplot as plt

# Average rewards for DQN and Q-Learning
dqn_avg_rewards = 220
q_learning_avg_rewards = 140

# Labels for the bars
labels = ['DQN', 'Q-Learning']

# Heights of the bars
heights = [dqn_avg_rewards, q_learning_avg_rewards]

# Colors for the bars
colors = ['blue', 'orange']

# Plotting the bar graph
plt.figure(figsize=(8, 6))
plt.bar(labels, heights, color=colors)
plt.xlabel('Models')
plt.ylabel('Average Rewards')
plt.title('Average Rewards Comparison: DQN vs Q-Learning')
plt.ylim(0, max(heights) * 1.2)  # Set y-axis limit slightly above the maximum reward
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Average vehicle wait times
wait_times = [25, 35]  # DQN, Q-Learning

# Throughput (vehicles per hour)
throughput = [1200, 1000]  # DQN, Q-Learning

# Traffic congestion levels (percentage reduction)
congestion_reduction = [30, 20]  # DQN, Q-Learning

# Metrics labels for the x-axis
metrics_labels = ['Avg Wait Time (s)', 'Throughput (vehicles/hr)', 'Congestion Reduction (%)']

# Plotting the comparative performance
plt.figure(figsize=(10, 6))

# Bar plot for average wait times
plt.subplot(2, 2, 1)
plt.bar(['DQN', 'Q-Learning'], wait_times, color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('Average Wait Time (s)')
plt.title('Average Vehicle Wait Times')

# Bar plot for throughput
plt.subplot(2, 2, 2)
plt.bar(['DQN', 'Q-Learning'], throughput, color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('Throughput (vehicles/hr)')
plt.title('Throughput')

# Bar plot for congestion reduction
plt.subplot(2, 2, 3)
plt.bar(['DQN', 'Q-Learning'], congestion_reduction, color=['blue', 'green'])
plt.xlabel('Model')
plt.ylabel('Congestion Reduction (%)')
plt.title('Traffic Congestion Levels')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Experimental results data
metrics_labels = ['Avg Wait Time (s)', 'Throughput (vehicles/hr)', 'Congestion Reduction (%)']
dqn_results = [25, 1200, 30]  # DQN model results
q_learning_results = [35, 1000, 20]  # Q-Learning model results

# Plotting the analysis of results
plt.figure(figsize=(10, 6))

# Line plot for DQN results
plt.plot(metrics_labels, dqn_results, marker='o', label='DQN', color='blue')

# Line plot for Q-Learning results
plt.plot(metrics_labels, q_learning_results, marker='s', label='Q-Learning', color='green')

# Add labels and title
plt.xlabel('Metrics')
plt.ylabel('Performance')
plt.title('Analysis of Results')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
